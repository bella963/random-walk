```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def path_poisson(num_paths, steps, T, xip):
    x = np.zeros([num_paths, steps+1])
    x_c = np.zeros([num_paths, steps+1])
    time = np.zeros([steps + 1])
    
    dt = T / float(steps)
    
    z = np.random.poisson(xip*dt, [num_paths, steps])
    
    for i in range(steps):
        x[:,i+1] = x[:,i] + z[:,i]
        x_c[:,i+1] = x_c[:,i] - xip*dt + z[:,i]
        time[i+1] = time[i] + dt
        
    path = {"time":time, "x":x, "x_c":x_c}
    
    return path
        
    
```

```python
num_paths = 30
steps = 500 #steps record the actual process
T = 30
xip = 1
    
P = path_poisson(num_paths,steps, T, xip)
time = P["time"]
x = P["x"]
xc = P["x_c"]
    
    
plt.figure(1)
plt.plot(time, np.transpose(x))
plt.xlabel("time")
plt.ylabel("x")
plt.show()
    
    
plt.figure(2)
plt.plot(time, np.transpose(xc),"-")
plt.xlabel("time")
plt.ylabel("x_c")
plt.show()
    
    
```
